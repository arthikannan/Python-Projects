```python
#groupby

dict = {"Company":["Google", "Facebook", "Amazon", "Google", "Google", "Amazon", "Facebook", "Amazon"],
        "People":["Sarah","Daniel","Kevin", "Courtney", "John", "Dennise", "Clara", "Hanna"],
        "Salary":[120,200,150,100,90,80,210,130]}
new_df = pd.DataFrame(dict)
```

```python
new_df
```

```python
new_df.groupby("Company").count()
```

```python
new_df.groupby("Company").describe().transpose()
```

```python
# Merging, joining and concatenating
df1 = pd.DataFrame(randn(4,3))
df2 = pd.DataFrame(randn(4,3))
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],1)
```

```python
df1[2]=(1,1,0,0)
df2[2]=(1,1,2,2)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,how="outer",on=2)
```

```python
df1.join(df2,how="outer", rsuffix="r")
```

```python
# dataframe operations

dict1 = {'A':["foo", "foo", "foo", "bar", "bar", "bar"],
         'B':["x","y","x","y","x","y"],
         'C':[1,2,3,1,5,3]}
dict1["D"] = [1,1,2,2,1,1]
df4 = pd.DataFrame(dict1)
```

```python
df4
```

```python
df4.pivot_table(values="C",index=["A","D"], columns="B")
```
